## Graph:
在了解BFS與DFS前，我認為要先了解什麼是Graph。

**Graph**叫做「圖」，是利用點與線來記錄點跟點之間的關聯，

不論每個點的形狀、大小、位置，每條線的粗細、長短、類型怎麼改變，

Graph只是為了記錄點跟點之間的關聯，所以上述點跟線的型態改變都沒關係，

要注意的是哪個點連哪個點一定不能改變。

上課提到Graph沒有強調每條線的方向性，但Graph可以在點跟點之間的線用箭頭表示方向，

叫做**Directed Graph**。

也可以在每條線加上權重(Weight)，用於後面上課的內容**Minimun Spanning Tree**


## BFS原理:
Breadth-First Search(BFS)，中文叫廣度優先搜尋，

是一種以Queue為基底的搜尋方式，

我們可以利用BFS將Graph像是Tree的形式分出每個level有哪幾個點。
### BFS的執行方法:
首先要先建兩個list，一個用來放入結果(以list2代稱)，一個用來放即將被放入list2的暫存list(以list1代稱)。

在list1中我們以**queue**的形式去做push跟pop，以下為執行步驟:

一開始先將開頭的點(EX:1)丟入list2中，並將跟1有連接的點以queue的方式push至list1中，

接著把list1中執行pop(把**最先**丟入list1的數刪掉)，並把pop的數加到list2中，再把pop掉的數所連接的數檢查是否有寫在list1 or list2中，

如果有重複就不再加入，如果沒有重複就push到list1中。

接著重複以上的步驟，直到list1中沒有數(表示所有數都被丟到list2中)，即完成。

**因為是以queue的方式進行，所以在list1中，level2的數一定接在level1後，level3的數一定接在level2後，以此類推**

**而也因為是以queue的方式進行，當list1的數要pop並加入到list2中時，一定也是先丟level1再丟level2，以此類推**

最後得到的list2就是依照level排列出來的list。
### BFS的時間複雜度(以 Adjacency List 為例):
(V是點的數量、E是邊的數量)

時間複雜度 : O(V+E)

## DFS原理:
Depth-First Search(DFS)，中文叫深度優先搜尋，

是一種以Stack為基底的搜尋方式，

我們可以利用DFS得到將所有點都跑過的路徑。

### DFS的執行方法:
一樣先建兩個list，一個用來放結果(以list2代稱)，一個用來放即將被放入list2的暫存list(以list1代稱)。

在list1中以**stack**的形式去做push跟pop，以下為執行步驟:

一開始先把開頭的點(EX:2)丟入list2中，並把跟2有連接的點以stack的方式push到list1中，

接著把list1中執行pop(把**最後**丟入list1的數刪掉)，並把pop掉的數加到list2中，再把pop掉的數所連接的數檢查是否有寫在list1 or list2中，

如果有重複就不再加入，如果沒有重數就push到list1中。

重複以上的步驟，直到list1中沒有數(表示所有數都被丟到list2中)，即完成。

**因為是以stack的方式進行，所以在list1中，每次pop掉的數與push進去的數一定是相連的**

**因為每次丟入list與丟出list1的數是相連的，所以漸漸加到list2中，最後list2會連成一條將每個點走過的路線**

### DFS的時間複雜度(以 Adjacency List 為例):
(V是點的數量、E是邊的數量)

時間複雜度 : O(V+E)   

## BFS與DFS比較:
BFS跟DFS主要有三個不同處:
1. 程式執行方面:

在BFS的部分是使用queue的方式去執行，我push完後要從list的**「最前面」**pop掉一個數到另一個list；

而在DFS則是用stack的方式進行，push完後要從list的**「最後面」**pop掉一個數到另一個list中。

2. 代表意涵不同:

兩個搜尋法雖然都會把每個點走過一次，但因為走的過程不同，代表的也不同。

在BFS當中，在起始點會先把跟那個點相連的點都走一次記為level1，再把跟level1相鄰的點走過，標記為level2，以此類推。

所以在最後出來的list會像是[起始點,level1,level2.....]。

而在DFS當中，在起始點會以先遇到的點作為新的起點，一直走下去，如果遇到死路就返回前面有其他相鄰節點的點，直到所有相鄰的點都被走過一次。

所以在最後出來的list會變成整個graph跑過一次的路徑(像完走迷宮的感覺)。

3. 使用情況不同:

在記憶空間的方面，因為BFS的記憶空間與點的數量呈正比，而DFS是跟遞迴深度成正比，

所以記憶空間DFS會比BFS小，如果是在很多點的情況下，會選擇使用DFS而不會選擇BFS。

## 學習歷程:
### BFS:
在BFS中我先去看了一開始寫的Queue的程式碼，並搭配老師上課講解BFS的過程，

將整個程式碼的架構在腦中先跑過了一次，

**問題1**

一開始一直在想該如何重複執行push的動作，因為之前有用過while也用過recursive，


後來因為我的BFS一開始會建兩個空的list，用recursive呼叫可能比較不適合，所以後來用while。

**問題2**

該如何判斷push進queue1中的數有沒有在queue1或queue2中重複，

我的想法是新建一個list來放已經丟入過的數，利用這個list去跟要丟入的數比較，看是否有重複。

一開始的概念如下:

(非完整程式碼)

```python
from collections import defaultdict 
class Graph:
    def __init__(self): 
        self.graph = defaultdict(list) 
    def addEdge(self,u,v): 
        self.graph[u].append(v)    
    def BFS(self, s): 
        queue1 =[]              #要怎麼持續丟進去(recursive(N)、while(Y))
        queue2 =[]              #一開始就丟進去的話，要在什麼時候去看queue1裡的數是否有在queue3裡面
        queue2.append(s)
        queue1.extend(self.graph[s])
        queue3=queue1+queue2
        j=0
        if len(queue1) != 0 :
            for i in queue1:
                while j <= len(queue3)-1:
                    if i !=queue3[j]:
                        j+=1
                    else:
                        break
                queue3=queue1+queue2
                queue2.append(i)
                queue1.extend(self.graph[i])
                queue1=queue1[1:]
                    
        return queue2
```

```python
g = Graph()
g.addEdge(1,2)
g.addEdge(1,4)
g.addEdge(2,3)
g.addEdge(2,6)
g.addEdge(3,5)
g.addEdge(3,7)
g.addEdge(7,5)
g.addEdge(5,2)
g.addEdge(5,6)
g.addEdge(6,1)
g.addEdge(4,6)
print(g.BFS(7))
```

我的想法是先建兩個空的list(queue1、queue2，用來提醒自己BFS是以queue的方式執行)，

如果queue1中還有東西，則持續動作直到queue1中沒有東西(全部加到queue2中)，

如果queue1中還有數，則利用for去跑每個數，並去和queue3比較看是否有相同(queue3放queue1和queue2中的數)，避免與前面放過的數重複。

如果沒有重複就丟到queue2中，並把這個數所連接的數加到queue1中。

**注意:這裡是用extend，在大一那年的程式設計課中有提到append跟extend的差別，**

**如果用append會把整個list丟到那個位子，用extend則會將list中的數依序放入另一個list中。**

### 錯誤:
這裡我有想到一個錯誤，就是我這樣只會把一開始queue1中原本就有的數丟到queue2中並把他們連接的數丟到queue1中，

但關於之後丟入queue1的數不會重複執行上面的動作。

```python
from collections import defaultdict 
class Graph:
    def __init__(self): 
        self.graph = defaultdict(list) 

    def addEdge(self,u,v): 
        self.graph[u].append(v) 
  
    def BFS(self, s): 
        """
        :type s: int
        :rtype: list
        """
        queue1 =[]                  
        queue2 =[]                  
        if len(queue2) == 0 :   
            queue2.append(s)
            queue1.extend(self.graph[s])    
        while len(queue1) != 0 :        #注意
            for i in queue1:
                queue3=queue1+queue2
                queue2.append(i)
                queue1=queue1[1:]
                for j in self.graph[i]:
                    if j not in queue3:
                        queue1.append(j)
        return queue2
```

為了解決可以讓後面加到queue1的數(i)也可以再讓self.graph[i]加進去，

我在最下面再寫一個丟入self.graph[i]的迴圈，

並把if的迴圈改成while(上方 #注意 處)，

這樣當我把新的數丟入queue1時就可以再繼續把數丟到queue2中。

並把最上面加了一個if條件，當一開始queue2中沒有東西的時候，把開頭(s)丟到queue2裡面，self.graph[s]丟到queue1中。

```python
g = Graph()
g.addEdge(1,2)
g.addEdge(1,4)
g.addEdge(2,3)
g.addEdge(2,6)
g.addEdge(3,5)
g.addEdge(3,7)
g.addEdge(7,5)
g.addEdge(5,2)
g.addEdge(5,6)
g.addEdge(6,1)
g.addEdge(4,6)
print(g.BFS(7))
```

### DFS:
我當初stack是用linked list的方式寫，queue是用array的方式寫，

DFS要用stack寫但因為概念跟BFS很像，我的DFS也就直接用array方式完成。

原本有聽說有一種雙向的array，但因為我沒學過，所以這裡把一般的array透過[::-1]顛倒來寫，幫助自己了解。

以下是我一開始的架構:

(非完整程式碼)

```python
from collections import defaultdict 
class Graph:
    def __init__(self): 
        self.graph = defaultdict(list) 
    def addEdge(self,u,v): 
        self.graph[u].append(v) 
    def DFS(self, s):
        """
        :type s: int
        :rtype: list
        """
        stack1=[]
        stack2=[]
        if len(stack2)==0:
            stack2.append(s)
            stack1.extend(self.graph[s])
        while len(stack1) != 0 :            
            stack1=stack1[::-1]         
            for i in stack1:
                stack3=stack1+stack2
                stack2.append(i)
                stack1=stack1[1:]
                for j in self.graph[i]:
                    if j not in stack3:
                        stack1.insert(0,j)
        return stack2
```

**錯誤1**

這裡一開始有出現一個錯誤:忘記寫stack1=stack1[1:]，

當我寫for i in stack1時，我for裡在stack1[0]的位置insert東西後，

當我for跑到index=1的位置一樣會是原本index=0的數。

(如下示範碼)

**錯誤2**

我為了讓self.graph[s][-1]能被pop出來我在while執行stack1=stack1[::-1]，

但我發現這樣我必須去看什麼時候裡面的兩個for會跑完(會再跑一次while)，

發現這樣會讓自己更複雜。

```python
a=[5,3,1]
for i in a:
    if len(a)<6:
        a.insert(0,1)
        print(i)
```

所以後來我覺得可能第一部份的for直接改指定用stack1[0]，

然後使用stack1=stack1[1:]去把第一個刪除，對我來說會比較好了解。

(更改後的程式碼如下)

```python
class Graph:
    def __init__(self): 
        self.graph = defaultdict(list) 
    def addEdge(self,u,v): 
        self.graph[u].append(v) 
    def DFS(self, s):
        """
        :type s: int
        :rtype: list
        """
        stack1=[]
        stack2=[]
        if len(stack2)==0:
            stack2.append(s)
            stack1.extend(self.graph[s])
        stack1=stack1[::-1]
        while len(stack1) != 0 :
            num = stack1[0]
            stack2.append(num)
            for i in self.graph[num]:
                stack3=stack1+stack2
                if i not in stack3:
                    stack1.insert(0,i)
            stack1=stack1[1:]
        return stack2
```

這裡會進infinty loop，

後來我重新看了一次，我認為是因為我在for裡面一直insert(0,i)，

但每加一個時如果沒有調整list裡面的狀況，會導致我的for迴圈一值進行下去(就如剛剛上面的示範碼，如果沒有設定if len(a)<6，也會進無限迴圈)

而且在最一開始要被pop掉的數不會被pop掉，因為照目前程式碼，會先執行完for，insert完才pop掉，刪掉的會是新進去的數。

所以我調整了一下我的stack1=stack1[1:]這句的位置。

```python
class Graph:
    def __init__(self): 
        self.graph = defaultdict(list) 
    def addEdge(self,u,v): 
        self.graph[u].append(v) 
    def DFS(self, s):
        """
        :type s: int
        :rtype: list
        """
        stack1=[]
        stack2=[]
        if len(stack2)==0:
            stack2.append(s)
            stack1.extend(self.graph[s])
        stack1=stack1[::-1]
        while len(stack1) != 0 :
            num = stack1[0]
            stack1=stack1[1:]
            stack2.append(num)
            for i in self.graph[num]:
                stack3=stack1+stack2
                if i not in stack3:
                    stack1.insert(0,i)
        return stack2
```

```python
g = Graph()
g.addEdge(1,2)
g.addEdge(1,4)
g.addEdge(2,3)
g.addEdge(2,6)
g.addEdge(3,5)
g.addEdge(3,7)
g.addEdge(7,5)
g.addEdge(5,2)
g.addEdge(5,6)
g.addEdge(6,1)
g.addEdge(4,6)
print(g.DFS(7))
```

### 成功了
DFS除了一些比較細微的錯誤要修改以外，
